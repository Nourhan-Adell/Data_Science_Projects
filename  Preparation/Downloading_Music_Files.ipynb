```python
# import all necessary libraries 
import youtube_dl # slow 
import pandas as pd
import yt_dlp
import pickle
# from future import unicode_literals
```

```python
#read in csv data 
import os 
pwd = os.getcwd()
csv_path = os.path.join(pwd, 'Billboard_Data.csv')
data = pd.read_csv('/content/Billboard.csv')
df = data.copy()
df.head()
```

```python
print(pwd)
```

```python
# Popular songs:
popular_song_titles = list(df.loc[df['Label']==1, 'Title']) # list of song titles on popular chart 
popular_artists = list(df.loc[df['Label']==1, 'Artist'])  #list for artists on popular chart
print('Length of popular songs: ', len(popular_song_titles))
print('Length of popular artists: ', len(popular_artists))
print()

# Unpopular songs:
unpopular_song_titles = list(df.loc[df['Label']==0, 'Title']) # list for songs on unpopular chart
unpopular_artists = list(df.loc[df['Label']==0, 'Artist'])  #list for artists on unpopular chart 
print('Length of unpopular songs: ', len(unpopular_song_titles))

print('Length of popular artists: ',len(unpopular_artists))
```

```python
popular_links_path = '/content/Downloaded_Songs/Popular_Songs'
unpopular_links_path = '/content/Downloaded_Songs/Unpopular_Songs'
```

```python
# helper function to download mp3 songs of specific song

def download_audio(song_name, artist, dir_name):
    dir = f'{dir_name}/{song_name}' # directory to store popular / unpopular songs
    filename = f'{dir}.mp3'
    ytdl_opts = {'format': 'bestaudio/best', # best audio format
                "outtmpl": filename, #output format with mp3 extension
                'noplaylist': True, # disable download playlist
                }
    song_info =  yt_dlp.YoutubeDL().extract_info(f"ytsearch:{song_name}-{artist}", download=False) # search for song with specific artist name
    with yt_dlp.YoutubeDL(ytdl_opts) as ydl:
        try:
            ydl.download([song_info['webpage_url']])
        except:
            print('could not download song')
```

## Download full Popular songs:

```python
for i in range(len(popular_song_titles)):
  download_audio(popular_song_titles[i], popular_artists[i],popular_links_path  )
```

```python
popular_downloaded_dir = os.path.join(pwd, 'Downloaded_Songs/Popular_Songs') # path to already downloaded popular songs
popular_songs_downloaded = os.listdir(popular_downloaded_dir) # list of already downloaded popular songs 
popular_songs_downloaded = [song.replace('.mp3','') for song in popular_songs_downloaded] # get song names from already downloaded list
len(popular_songs_downloaded)
```

## Download full Unpopular songs:

```python
for i in range(len(unpopular_song_titles)):
  download_audio(unpopular_song_titles[i], unpopular_artists[i], unpopular_links_path  )
```

```python
unpopular_downloaded_dir = os.path.join(pwd, 'Downloaded_Songs/Unpopular_Songs') # path to already downloaded unpopular songs
unpopular_songs_downloaded = os.listdir(unpopular_downloaded_dir) # list of already downloaded unpopular songs 
unpopular_songs_downloaded = [song.replace('.mp3','') for song in unpopular_songs_downloaded] # get song names from already downloaded list
len(unpopular_songs_downloaded)
```

```python
#Check on all the songs are already exist
for song, artist in zip(popular_song_titles[:], popular_artists[:]): # iterate through the popular songs to download 
    if song not in popular_songs_downloaded: # check if song is already downloaded 
        #print(song)
        download_audio(song ,artist, popular_links_path)
    else:
        print('song already exists')
```

```python
song_path = [] # list to contain path of donloaded popular & unpopular songs 
for song in popular_song_titles:
    if song in popular_songs_downloaded:
        song_path.append(f'popular/{song}.mp3') # add popular downloaded songs to list 
    else:
        song_path.append(f'n/a')
```

```python
for song in unpopular_song_titles:
    if song in unpopular_songs_downloaded:
        song_path.append(f'unpopular/{song}.mp3') # add unpopular downloaded songs to song path list 
    else:
        song_path.append(f'n/a')
```
